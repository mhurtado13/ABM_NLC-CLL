```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

##### Load experimental data

```python
experimental = np.loadtxt('../../Netlogo_NLCs-CLL/filtered_fused_9patients.csv', delimiter=",", skiprows=1)
viability_exp = experimental[:,1]
concentration_exp = experimental[:,2]
time_points = experimental[:,0]
```

##### Load output from simulation

```python
viability = pd.read_csv('../data_output/viability.csv', delimiter=",")
concentration = pd.read_csv('../data_output/concentration.csv', delimiter=",")
```

### Uptake rate cancer

```python
param_name = "uptake_rate_cancer"
vals = [0, 2, 3, 5, 7, 9]
```

#### Viability

```python
# Iterate over each column of the matrix
for i in range(viability.shape[1]):
    # Plot each column as a separate line, using a different color for each line
    plt.plot(time_points, viability.iloc[:, i], label=f'Value = {vals[i]}')

plt.plot(time_points, viability_exp, label='Experimental', color = 'black')
plt.xlabel('Time')
plt.ylabel('Viability')
plt.title(f'Viability: {param_name}')
plt.legend(loc='upper right', bbox_to_anchor=(1.35, 0.5))

plt.grid(True)
plt.show()
```

#### Concentration

```python
# Iterate over each column of the matrix
for i in range(concentration.shape[1]):
    # Plot each column as a separate line, using a different color for each line
    plt.plot(time_points, concentration.iloc[:, i], label=f'Value = {vals[i]}')

plt.plot(time_points, concentration_exp, label='Experimental', color = 'black')
# Add labels and legend
plt.xlabel('Time')
plt.ylabel('Concentration')
plt.title(f'Concentration: {param_name}')
plt.legend(loc='upper right', bbox_to_anchor=(1.35, 0.5))

plt.grid(True)
plt.show()
```
